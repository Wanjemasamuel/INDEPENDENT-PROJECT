```python
import pandas as pd
import statistics as std
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as py
```

```python
# loading the dataset
df=pd.read_csv('Financial Statements.csv', encoding='latin-1')
df
```

```python
#getting an overview summary of the dataset inclusively
df.describe(include='all')
```

```python
#understanding how the data is structured interms of the data types per column
df.info()
```

```python
#converting Earning per share, Revenue and ROI data types from float64 to int64
df['Earning Per Share'] = df['Earning Per Share'].astype('int64')
df['Revenue'] = df['Revenue'].astype('int64')
df['ROI'] = df['ROI'].astype('int64')
df
```

```python
#loading the first ten rows of the dataset
df.head(10)
```

```python
#loading the last ten rows of the dataset
df.tail(10)
```

```python
# Check for missing values in the entire DataFrame
missing_values = df.isnull()
print(missing_values)
```

```python
#this is to count all the Null values within the whole Dataset
df.count()
df
```

```python
#This is counting the Null value within the specified columns( Revenue, Earning per Share and ROI) in the dataset
df[['Revenue', 'Earning Per Share', 'ROI']].count()
```

```python
# Drop null values from specific columns
cleaned_df = df.dropna(subset=['Revenue', 'Earning Per Share', 'ROI'])
cleaned_df
```

```python
#Before replacement
#Pivot table that just represent how per category the Revenues compare and the ROI which is return on Investment
pivot_df = pd.pivot_table(cleaned_df, index='Category', values=['Revenue', 'ROI'])
pivot_df
```

```python
# Replace 'BANK' with 'Bank' in the Category column
cleaned_df['Category'] =cleaned_df['Category'].replace('BANK', 'Bank')
cleaned_df
```

```python
#After replacement of BANK to Bank
#Pivot table that just represent how per category the Revenues compare and the ROI which is return on Investment
pivot_cleaned_df = pd.pivot_table(cleaned_df, index='Category', values=['Revenue', 'ROI'])
pivot_cleaned_df
```

```python
#Data Visualisation
#Comparing Earning Per Share, Revenue and ROI Using various graph representations
```

```python
# Plot the Earning Per Shares values
#Assuming 'cleaned_df' is your DataFrame containing the Earning Per Share data
plt.figure(figsize=(8,10))
sns.barplot(x='Category', y='Earning Per Share', data=cleaned_df)
plt.xlabel('Category')
plt.ylabel('Earning Per Share')
plt.title('Earning Per Share')
plt.show()
```

```python
#Revenue Graphical representation
plt.figure(figsize=(10,6))
sns.boxplot(x='Category', y='Revenue', data=cleaned_df)
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.title('Revenue')
plt.show()
```

```python
#ROI graphical representation
plt.figure(figsize=(10, 6))
plt.plot(cleaned_df['Category'], cleaned_df['ROI'], marker='o', linestyle='-')
plt.xlabel('Category')
plt.ylabel('ROI')
plt.title('ROI Over Time')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(True) 
plt.show()
```

```python
#Scatter plot to show correlation between revenue and ROI
plt.figure(figsize=(8, 6))
plt.scatter(df['Earning Per Share'], df['Revenue'], alpha=0.5)
plt.title('Scatter Plot of Revenue vs ROI')
plt.xlabel('Earning Per Share')
plt.ylabel('Revenue')
plt.show()
```

```python
#HeatMAP showing the relationship between the three column items
correlation_matrix = df[['Revenue', 'ROI', 'Total Debt']].corr()

plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Heatmap Showing Relationship between Revenue, ROI and Total Debt')
plt.show()
```
